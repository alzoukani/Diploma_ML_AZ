```python
import numpy as np
import pandas as pd
from google.colab import files
t = np.linspace(0,60,100)
Wmax = 1.97
k = 0.05
w_clean = Wmax*(1-np.exp(-k*t))
np.random.seed(42)
noise_std = 0.015
noise = np.random.normal(0, noise_std,size = len(w_clean))
w_noisy = w_clean + noise
w_noisy[0] = 0.0
w_noisy = np.maximum.accumulate(w_noisy)
w_noisy = np.clip(w_noisy, 0, Wmax)
df = pd.DataFrame({"Time_days": t, "Produced_Water_cc": w_noisy})
df.to_excel("synthetic_water_production.xlsx", index=False)
files.download("synthetic_water_production.xlsx")
```
