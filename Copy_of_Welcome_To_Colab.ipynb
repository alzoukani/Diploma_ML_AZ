```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic data
time = np.linspace(0, 120, 100)  # Time in minutes
mass_deionized = 0.005 * time  # Mass increase due to deionized water only

# Simulate a breakthrough after 60 minutes with the tracer being heavier
mass_tracer = mass_deionized + 0.01 * (time - 60) * (time > 60)

# Convert to DataFrame for easier manipulation
data = pd.DataFrame({'time': time, 'mass_deionized': mass_deionized, 'mass_tracer': mass_tracer})

# Smooth the data using a rolling average
data['mass_deionized_smooth'] = data['mass_deionized'].rolling(window=5).mean()
data['mass_tracer_smooth'] = data['mass_tracer'].rolling(window=5).mean()

# Drop NaN values that result from rolling
data.dropna(inplace=True)

# Update variables with smoothed data
time = data['time'].values
mass_deionized_smooth = data['mass_deionized_smooth'].values
mass_tracer_smooth = data['mass_tracer_smooth'].values

# Plot both masses to visually identify breakthrough
plt.plot(time, mass_deionized_smooth, label="Deionized Water (Baseline)")
plt.plot(time, mass_tracer_smooth, label="200 kppm NaCl (Tracer)", linestyle="--")

# Automatic breakthrough detection based on significant deviation
mass_diff = mass_tracer_smooth - mass_deionized_smooth
threshold = np.mean(mass_diff[:breakthrough_index]) + 2 * np.std(mass_diff[:breakthrough_index])  # Example threshold
breakthrough_index = np.argmax(mass_diff > threshold)
breakthrough_time = time[breakthrough_index]

# Plotting
plt.axvline(x=breakthrough_time, color="r", linestyle=":", label="Tracer Breakthrough")
plt.xlabel('Time (minutes)')
plt.ylabel('Mass on Balance (g)')
plt.title('Mass Accumulation vs Time during Tracer Injection')
plt.legend()
plt.grid(True)
plt.show()
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic data
time = np.linspace(0, 120, 100)  # Time in minutes
mass_deionized = 0.005 * time  # Mass increase due to deionized water only

# Simulate a breakthrough after 60 minutes with the tracer being heavier
mass_tracer = mass_deionized + 0.01 * (time - 60) * (time > 60)

# Convert to DataFrame for easier manipulation
data = pd.DataFrame({'time': time, 'mass_deionized': mass_deionized, 'mass_tracer': mass_tracer})

# Smooth the data using a rolling average
data['mass_deionized_smooth'] = data['mass_deionized'].rolling(window=5).mean()
data['mass_tracer_smooth'] = data['mass_tracer'].rolling(window=5).mean()

# Drop NaN values that result from rolling
data = data.dropna()

# Update variables with smoothed data
time_smooth = data['time'].values
mass_deionized_smooth = data['mass_deionized_smooth'].values
mass_tracer_smooth = data['mass_tracer_smooth'].values

# Calculate the difference and detect breakthrough point
mass_diff = mass_tracer_smooth - mass_deionized_smooth
breakthrough_index = np.argmax(mass_diff > 0)  # Detect where tracer starts to significantly add mass
breakthrough_time = time_smooth[breakthrough_index]
breakthrough_mass = mass_tracer_smooth[breakthrough_index]

# Print the breakthrough time and value
print(f"Breakthrough Time: {breakthrough_time} minutes")
print(f"Breakthrough Mass: {breakthrough_mass} grams")

# Plotting
plt.plot(time_smooth, mass_deionized_smooth, label='Deionized Water (Baseline)')
plt.plot(time_smooth, mass_tracer_smooth, label='200 kppm NaCl (Tracer)', linestyle='--')
plt.axvline(x=breakthrough_time, color="r", linestyle=":", label="Tracer Breakthrough")
plt.xlabel('Time (minutes)')
plt.ylabel('Mass on Balance (g)')
plt.title('Mass Accumulation vs Time during Tracer Injection')
plt.legend()
plt.grid(True)
plt.show()



```
